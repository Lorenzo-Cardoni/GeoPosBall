```python
import pandas as pd
import numpy as np
import folium
import json
```

```python
from pyproj import Proj

zone_number = 33

def convert_dd_to_utm(latitude, longitude):

    if latitude == 0 or longitude == 0:
        return 0,0

    # Definisci la proiezione UTM
    utm_proj = Proj(proj='utm', zone=zone_number, ellps='WGS84')

    # Effettua la conversione delle coordinate
    utm_easting, utm_northing = utm_proj(longitude, latitude)

    return utm_easting, utm_northing
```

```python
mean_1_ll = [np.mean(df_1["latitude_2"]), np.mean(df_1["longitude_2"])]
median_1_ll = [ np.median(df_1["latitude_2"]), np.median(df_1["longitude_2"])]

east,north = convert_dd_to_utm(mean_1_ll[0],mean_1_ll[1])
mean_1_utm = [east,north]
east,north = convert_dd_to_utm(median_1_ll[0],median_1_ll[1])
median_1_utm = [east,north]
variance_1_utm = [ np.var(df_1["utm_east"]), np.var(df_1["utm_north"]) ]

pdop_mean_1 = np.mean(df_1["PDOP"])

map_1 = folium.Map(location=[df_1["latitude_2"].iloc[0], df_1["longitude_2"].iloc[0]], zoom_start=18)

#for x in zip(df_2["latitude_2"].to_list(), df_1["longitude_2"].to_list()):
  #print(x)
#  folium.Marker(location = x, popup = x).add_to(map_1)

folium.Marker(location = mean_1_ll, tooltip= "mean", icon=folium.Icon(color="red")).add_to(map_1)
folium.Marker(location = median_1_ll, tooltip= "median", icon=folium.Icon(color="black")).add_to(map_1)


map_1
```
