```python
import numpy as np
import pandas as pd
```

```python
col = ["time","acc_x","acc_y","acc_z","gyro_x","gyro_y","gyro_z","angle_1",
       "angle_2","angle_3", "magn_1", "magn_2", "magn_3", "pressure", "altitude",
       "Dstatus_1","Dstatus_2","Dstatus_3","Dstatus_4", "longitude", "latitude",
       "longitude_2","latitude_2","GPS_height(m)","GPSYaw(Deg)", "GPSV(km/h)", "SN", "PDOP", "HDOP", "VDOP"]
df = pd.DataFrame(columns=col)
```

```python
#qua metto tutti i risultati poi li aggiungo come colonne al df alla fine
times = []
acc_x = []
acc_y = []
acc_z = []
gyro_x = []
gyro_y = []
gyro_z = []
angle_1 = []
angle_2 = []
angle_3 = []
magn_1 = []
magn_2 = []
magn_3 = []
pressure = []
altitude = []
Dstatus_1 = []
Dstatus_2 = []
Dstatus_3 = []
Dstatus_4 = []
longitudine = []
latitudine = []
longitudine_2 = []
latitudine_2 = []
GPS_height = []
GPSYaw = []
GPSV = []
SN = []
PDOP = []
HDOP = []
VDOP = []
```

```python
def parse_time(values : list):
  times.append(values[0].strip())
#  times.append(':'.join(values))

def parse_Acc(values : list):
  val = values[0].split(' ')
  acc_x.append(val[0])
  acc_y.append(val[1])
  acc_z.append(val[2])
  #print(acc_x)

def parse_gyro(values : list):
  val = values[0].split(' ')
  gyro_x.append(val[0])
  gyro_y.append(val[1])
  gyro_z.append(val[2])

def parse_angle(values : list):
  val = values[0].split(' ')
  angle_1.append(val[0])
  angle_2.append(val[1])
  angle_3.append(val[2])

def parse_magnet(values:list):
  val = values[0].split(' ')
  magn_1.append(val[0])
  magn_2.append(val[1])
  magn_3.append(val[2])

#pressione e altitudine
def parse_pressure(values : list):
  val = values[0].split(' ')
  pressure.append(val[0])
  altitude.append(val[1])

def parse_Dstatus(values : list):
  val = values[0].split(' ')
  Dstatus_1.append(val[0])
  Dstatus_2.append(val[1])
  Dstatus_3.append(val[2])
  Dstatus_4.append(val[2])

def parse_longitude(values : list): # e latitudine
  val1 = values[0].split(' ')
  longitudine.append(val1[0])
  latitudine.append(values[1])

def parse_longitude_2(values : list):
  longitudine_2.append(values[0].strip())

def parse_latitudine_2(values : list):
  latitudine_2.append(values[0].strip())

def parse_GPSHeight(values : list): # e gps_yaw, gps_v
  val1 = values[0].split(' ')
  GPS_height.append(val1[0].replace("m",""))
  val2 = values[1].split(' ')
  GPSYaw.append(val2[0].replace("Deg",""))
  GPSV.append(values[2].replace("km/h",""))

def parse_SN(values : list): # e PDOP, HDOP, VDOP
  val1 = values[0].split(' ')
  SN.append(val1[0])
  val2 = values[1].split(' ')
  PDOP.append(val2[0])
  val3 = values[2].split(' ')
  HDOP.append(val3[0])
  VDOP.append(values[3])

```

```python
counter = 0
with open("/content/Dati_misa_posizione_2_24112023.txt", 'r') as file:
  for line in file:
    line = line.rstrip()
    splt_line = line.split(':')
    if(len(splt_line) > 1):
      etichetta = splt_line[0]
      values = splt_line[1:]
      #print(etichetta,values)
      if(etichetta == "Time"):
        parse_time(values)
      elif(etichetta == "Acc"):
        parse_Acc(values)
      elif(etichetta == "Gyro"):
        parse_gyro(values)
      elif(etichetta == "Angle"):
        parse_angle(values)
      elif(etichetta == "Mag"):
        parse_magnet(values)
      elif(etichetta == "Pressure"):
        parse_pressure(values)
      elif(etichetta == "DStatus"):
        parse_Dstatus(values)
      elif(etichetta == "Longitude"):
        parse_longitude(values)
      elif(etichetta == "Longitude_2"):
        parse_longitude_2(values)
      elif(etichetta == "Latitude_2"):
        parse_latitudine_2(values)
      elif(etichetta == "GPSHeight"):
        parse_GPSHeight(values)
      elif(etichetta == "SN"):
        parse_SN(values)
      else:
        print("problemmm");
```
